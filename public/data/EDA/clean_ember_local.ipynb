# Limpieza in-place — Ember (local)
Lee y sobrescribe `yearly_full_release_long_format.csv`.
También guarda `ember_tidy.csv` en este directorio.

```python
import pandas as pd, os
from pathlib import Path
IN_FILE = 'yearly_full_release_long_format.csv'
OUT_TIDY = 'ember_tidy.csv'
if not os.path.isfile(IN_FILE):
    raise FileNotFoundError(f'No se encontró {IN_FILE}. Coloca el CSV en este directorio.')
def normalizar_decimales(df, cols):
    for c in cols:
        if c in df.columns:
            df[c] = (
                df[c].astype(str)
                .str.replace(r'[^0-9eE.,\-]', '', regex=True)
                .str.replace(',', '.', regex=False)
            )
            df[c] = pd.to_numeric(df[c], errors='coerce')
    return df
df = pd.read_csv(IN_FILE)
df.columns = (df.columns
    .str.strip()
    .str.replace(' ', '_')
    .str.replace(r'[^0-9a-zA-Z_]', '', regex=True)
    .str.lower())
df = df.rename(columns={'iso3':'iso_3_code','iso_3':'iso_3_code','iso_3country_code':'iso_3_code','iso_3_country_code':'iso_3_code'})
df = normalizar_decimales(df, ['year','value'])
df = df.dropna(subset=['year','value','area','category','variable']).copy()
df['year'] = df['year'].astype(int)
df.to_csv(IN_FILE, index=False)
df.to_csv(OUT_TIDY, index=False)
print('Sobrescrito →', IN_FILE)
print('Tidy →', OUT_TIDY)
df.describe(include='all')
```
